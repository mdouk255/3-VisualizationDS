```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r'C:\Users\19294\Downloads'
os.chdir(path)
df = pd.read_csv('Large_NYC_Visualization_Dataset.csv',sep = ',')
df.head()
```

```python
df.columns
```

```python
corr_matrix = df[['Average_Temperature', 'Total_Sales_Thousands', 'Population_Thousands', 'House_Prices_Median', 'Unemployment_Rate']].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.boxplot(df['House_Prices_Median'])
plt.ylabel('Median House Price')
plt.title('Box Plot of Median House Prices')
plt.show()
```

```python
grouped_df = df.groupby('Neighborhood').agg({'House_Prices_Median': 'mean', 'Unemployment_Rate': 'mean'})
plt.figure(figsize=(10, 6))

# Plot the average median house prices
plt.bar(grouped_df.index, grouped_df['House_Prices_Median'], label='Average House Price')

# Plot the average unemployment rates
plt.bar(grouped_df.index, grouped_df['Unemployment_Rate'], label='Average Unemployment Rate')

# Customize the plot
plt.xlabel('Neighborhood')
plt.ylabel('Value')
plt.title('Average House Prices and Unemployment Rates by Neighborhood')
plt.xticks(rotation=45)
plt.legend()

# Show the plot
plt.show()
```
